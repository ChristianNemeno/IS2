```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

file_path = 'water_potability_master_clean.xlsx'
df = pd.read_excel(file_path)


# 1. Count the raw numbers
counts = df['Potability'].value_counts()
print("Counts:\n", counts)

# 2. Check the percentage (The most important part)
percentages = df['Potability'].value_counts(normalize=True) * 100
print("\nPercentages:\n", percentages)

# 3. Visualize it
plt.figure(figsize=(6, 4))
sns.countplot(x='Potability', data=df, palette='pastel')
plt.title('Class Balance: 0 (Not Potable) vs 1 (Potable)')
plt.show()
```
